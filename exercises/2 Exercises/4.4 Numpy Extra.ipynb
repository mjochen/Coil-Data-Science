# Exercises numpy

Numpy is embedded quite well into pandas (which is the next chapter). NumPy is very useful for performing logical and mathematical calculations on arrays and matrices. This tool performs these operations much faster and more efficiently than Python lists. 

Furthermore, Numpy uses less memory and storage space, which is the main advantage. This is because numpy is typed. Normal Python is typeless, so if you create a list of integers, Python ensures that it is also suitable for floating point numbers or very long strings. The more items you add to the list, the more space is used as a list grows exponentially. This means that a lot of memory is required, whereas numpy knows what the array is intended for and allocates exactly enough memory.
